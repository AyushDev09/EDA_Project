# ZOMATO DATASET EXPLORATORY DATA ANALYSIS

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('zomato.csv',encoding='latin-1')
df.head()
```

```python
# showing all the columns

df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
# Checking for null values

df.isnull().sum()
```

```python
# list comprehension for checking null values

[column for column in df.columns if df[column].isnull().sum() > 0]
```

```python
# Using a heatmap to visualize the null values
# Since there are 9551 rows and null values are 9, they are not visible on the heatmap

sns.heatmap(df.isnull(), yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
# combining both dataframes on country code using a left join

final_df = pd.merge(df,df_country,on = "Country Code", how = "left" )
```

```python
final_df.head()
```

```python
# number of records

final_df.Country.value_counts()
```

```python
# Index

country_name = final_df.Country.value_counts().index
```

```python
# Values

country_val = final_df.Country.value_counts().values
```

```python
# Pie chart for top 3 countries 

plt.pie(country_val[:3], labels = country_name[:3],shadow = True, autopct = '%1.2f%%')
```

Observation : Maximum records or transactions are from India, United States of America and United Kingdom coming in second and third
respectively

```python
# grouping by various rating columns

ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
# A bar plot to visualize 

plt.rcParams['figure.figsize'] = (12,6) # to increase figure size

sns.barplot(x = 'Aggregate rating',y = 'Rating Count', data = ratings, hue = 'Rating color',palette=['white','red','orange','yellow','green','green'])
```

```python
# finding out which country has given the most 0 rating

final_df.groupby(['Aggregate rating','Country']).size().reset_index().head(4)
```

```python
# finding out what currency is used by a country

final_df.groupby(['Country','Currency']).size().reset_index()
```

```python
final_df.City.value_counts()
```

```python
# Creating a piechart to visualize top 5 cities by distribution

city_values = final_df.City.value_counts().values
city_labels = final_df.City.value_counts().index
plt.pie(city_values[:5], labels = city_labels[:5], autopct = '%1.2f%%')
```
